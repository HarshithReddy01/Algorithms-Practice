```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

inputs = ["MRI", "Ultrasound", "Hormones", "Genomics", "Clinical History"]
ai_core = "AI Fusion Brain"
outputs = ["Diagnosis", "Risk Prediction", "Therapy Response"]

for i in inputs:
    G.add_edge(i, ai_core)

for o in outputs:
    G.add_edge(ai_core, o)

pos = nx.spring_layout(G, center=(0,0))

colors = []
for node in G:
    if node in inputs:
        colors.append("#FFA500")
    elif node == ai_core:
        colors.append("#8A2BE2")
    else:
        colors.append("#2ECC71")

plt.figure(figsize=(8,6))
nx.draw(G, pos, with_labels=True, node_color=colors, node_size=2000, font_size=10)
plt.title("Multimodal AI Fusion in Endocrinology", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import FancyBboxPatch
```

```python
# ----------------------------
# Master "AI in Endocrinology" System Diagram
# ----------------------------

G = nx.DiGraph()

# --- Layers / Modules ---
data_layer = [
    "Imaging\n(MRI/US/CT/PET)",
    "Labs & Hormones\n(TSH, cortisol, glucose)",
    "EHR + Clinical Notes\n(NLP-ready)",
    "Omics\n(genomics/proteomics)",
    "Wearables & CGM\n(time series)"
]

ai_core = [
    "Representation Learning\n(feature extraction)",
    "Multimodal Fusion\n(join signals)",
    "Prediction Engines\n(ML/DL/RL)",
    "Explainability\n(SHAP/Grad-CAM/XAI)"
]

disease_modules = [
    "Diabetes\n(screening  risk  therapy)",
    "Thyroid\n(nodules  cancer  dosing)",
    "Adrenal\n(incidentaloma  PA  Cushing)",
    "Pituitary\n(radiomics  CDS  surgery)",
    "PCOS\n(phenotyping  risk)",
    "Rare Endocrine Tumors\n(NETs  grading  survival)"
]

clinical_outcomes = [
    "Early Detection",
    "Risk Stratification",
    "Personalized Treatment",
    "Monitoring & Follow-up",
    "Resource Optimization\n(workflow)"
]

trust_ring = [
    "Bias & Equity\n(hidden stratification)",
    "Generalizability\n(distribution shift)",
    "Privacy\n(federated learning)",
    "Regulation\n(SaMD/EU AI Act)",
    "Post-deployment Surveillance\n(model drift)",
    "Clinical Workflow Fit\n(integration barriers)"
]

# --- Add nodes ---
for n in data_layer + ai_core + disease_modules + clinical_outcomes + trust_ring:
    G.add_node(n)

# --- Connect Data -> AI Core ---
for d in data_layer:
    G.add_edge(d, "Representation Learning\n(feature extraction)")
    G.add_edge(d, "Multimodal Fusion\n(join signals)")

G.add_edge("Representation Learning\n(feature extraction)", "Prediction Engines\n(ML/DL/RL)")
G.add_edge("Multimodal Fusion\n(join signals)", "Prediction Engines\n(ML/DL/RL)")
G.add_edge("Prediction Engines\n(ML/DL/RL)", "Explainability\n(SHAP/Grad-CAM/XAI)")

# --- AI Core -> Disease Modules ---
for m in disease_modules:
    G.add_edge("Prediction Engines\n(ML/DL/RL)", m)
    G.add_edge("Explainability\n(SHAP/Grad-CAM/XAI)", m)

# --- Disease Modules -> Clinical Outcomes ---
for m in disease_modules:
    for o in clinical_outcomes:
        # Keep diagram readable: only connect each module to a subset of outcomes
        if ("Diabetes" in m and o in ["Early Detection", "Risk Stratification", "Personalized Treatment", "Monitoring & Follow-up"]):
            G.add_edge(m, o)
        if ("Thyroid" in m and o in ["Early Detection", "Risk Stratification", "Personalized Treatment"]):
            G.add_edge(m, o)
        if ("Adrenal" in m and o in ["Early Detection", "Risk Stratification", "Monitoring & Follow-up"]):
            G.add_edge(m, o)
        if ("Pituitary" in m and o in ["Early Detection", "Risk Stratification", "Personalized Treatment", "Monitoring & Follow-up"]):
            G.add_edge(m, o)
        if ("PCOS" in m and o in ["Early Detection", "Risk Stratification", "Personalized Treatment"]):
            G.add_edge(m, o)
        if ("Rare Endocrine Tumors" in m and o in ["Early Detection", "Risk Stratification", "Monitoring & Follow-up"]):
            G.add_edge(m, o)

# --- Trust ring connects back into AI core + outcomes (governance loop) ---
for t in trust_ring:
    G.add_edge(t, "Prediction Engines\n(ML/DL/RL)")
    G.add_edge(t, "Resource Optimization\n(workflow)")

# ----------------------------
# Layout: Custom coordinates (for a clean, publication look)
# ----------------------------
pos = {}

# x positions for layers
x_data, x_ai, x_dis, x_out = 0.05, 0.35, 0.63, 0.90

# Spread nodes vertically per layer
def place_vertically(nodes, x, top=0.92, bottom=0.08):
    if len(nodes) == 1:
        return {nodes[0]: (x, (top+bottom)/2)}
    ys = [top - i*(top-bottom)/(len(nodes)-1) for i in range(len(nodes))]
    return {n: (x, y) for n, y in zip(nodes, ys)}

pos.update(place_vertically(data_layer, x_data))
pos.update(place_vertically(ai_core, x_ai, top=0.82, bottom=0.18))
pos.update(place_vertically(disease_modules, x_dis))
pos.update(place_vertically(clinical_outcomes, x_out, top=0.82, bottom=0.18))

# Trust ring (arranged around)
ring_coords = [
    (0.50, 0.98),
    (0.78, 0.92),
    (0.95, 0.55),
    (0.78, 0.18),
    (0.50, 0.02),
    (0.22, 0.18),
]
for node, (rx, ry) in zip(trust_ring, ring_coords):
    pos[node] = (rx, ry)

# ----------------------------
# Styling
# ----------------------------
colors = {}
for n in data_layer:
    colors[n] = "#FFA500"   # imaging/data vibe
for n in ai_core:
    colors[n] = "#8A2BE2"   # AI purple
for n in disease_modules:
    colors[n] = "#00B3B3"   # teal disease modules
for n in clinical_outcomes:
    colors[n] = "#2ECC71"   # green outcomes
for n in trust_ring:
    colors[n] = "#FF5A5F"   # red trust/risk ring

node_colors = [colors[n] for n in G.nodes()]

# ----------------------------
# Draw
# ----------------------------
plt.figure(figsize=(18, 10))
ax = plt.gca()
ax.set_facecolor("white")

# Background "group" panels (nice journal look)
def panel(x0, y0, w, h, label, fc, alpha=0.08):
    p = FancyBboxPatch((x0, y0), w, h,
                      boxstyle="round,pad=0.012,rounding_size=0.02",
                      linewidth=1.2, edgecolor=fc, facecolor=fc, alpha=alpha,
                      transform=ax.transAxes)
    ax.add_patch(p)
    ax.text(x0 + 0.01, y0 + h - 0.04, label, transform=ax.transAxes,
            fontsize=12, fontweight="bold", color=fc)

panel(0.01, 0.05, 0.22, 0.90, "DATA SOURCES", "#FFA500")
panel(0.25, 0.12, 0.22, 0.76, "AI CORE", "#8A2BE2")
panel(0.49, 0.05, 0.25, 0.90, "DISEASE MODULES", "#00B3B3")
panel(0.76, 0.12, 0.23, 0.76, "CLINICAL OUTCOMES", "#2ECC71")

# Edges
nx.draw_networkx_edges(
    G, pos,
    arrows=True,
    arrowstyle="-|>",
    arrowsize=12,
    width=1.0,
    alpha=0.45
)

# Nodes
nx.draw_networkx_nodes(
    G, pos,
    node_color=node_colors,
    node_size=2100,
    linewidths=1.2,
    edgecolors="#222"
)

# Labels
nx.draw_networkx_labels(
    G, pos,
    font_size=9,
    font_color="#111"
)

plt.title("Master System Diagram: AI-Driven Endocrine Intelligence (Multimodal  Models  Modules  Decisions + Trust Loop)",
          fontsize=14, fontweight="bold")

plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

# ---------- helper functions ----------
def add_box(ax, x, y, w, h, text, facecolor, fontsize=14, lw=2.5):
    box = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.03,rounding_size=0.04",
        linewidth=lw, edgecolor="black", facecolor=facecolor
    )
    ax.add_patch(box)
    ax.text(
        x + w/2, y + h/2, text,
        ha="center", va="center",
        fontsize=fontsize, fontweight="bold", color="black",
        linespacing=1.25
    )

def add_arrow(ax, x1, y1, x2, y2, text=None, fontsize=12):
    arr = FancyArrowPatch(
        (x1, y1), (x2, y2),
        arrowstyle="-|>", mutation_scale=18,
        linewidth=2.2, color="black"
    )
    ax.add_patch(arr)
    if text:
        ax.text(
            (x1+x2)/2, (y1+y2)/2 + 0.06, text,
            ha="center", va="bottom",
            fontsize=fontsize, fontweight="bold"
        )

# ---------- figure ----------
# Big canvas + high DPI = crisp text
fig, ax = plt.subplots(figsize=(18, 4.2), dpi=300)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

# ---------- box contents ----------
left_text = (
    "Multimodal Patient Data\n"
    " Imaging (MRI, US, CT, PET)\n"
    " Hormones & Labs\n"
    " EHR & Clinical Notes\n"
    " Omics\n"
    " Wearables"
)

core_text = (
    "AI Intelligence Core\n"
    " Feature Learning\n"
    " Multimodal Fusion\n"
    " Deep/ML Models\n"
    " Explainable AI"
)

models_text = (
    "Endocrine Disease Models\n"
    " Diabetes\n"
    " Thyroid Disorders\n"
    " Adrenal Disorders\n"
    " Pituitary Tumors\n"
    " PCOS\n"
    " Rare Endocrine Tumors"
)

clinical_text = (
    "Clinical Intelligence\n"
    " Early Detection\n"
    " Risk Stratification\n"
    " Personalized Therapy\n"
    " Continuous Monitoring"
)

trust_text = (
    "Trust & Governance Layer\n"
    " Bias Control\n"
    " Generalization\n"
    " Privacy (Federated Learning)\n"
    " Regulation\n"
    " Model Monitoring"
)

# ---------- layout (x, y, w, h) ----------
add_box(ax, 0.02, 0.20, 0.22, 0.60, left_text,  facecolor="#F5A623", fontsize=14)
add_box(ax, 0.27, 0.28, 0.18, 0.44, core_text,  facecolor="#6C63FF", fontsize=13)
add_box(ax, 0.49, 0.22, 0.20, 0.56, models_text, facecolor="#26A69A", fontsize=13)
add_box(ax, 0.72, 0.28, 0.18, 0.44, clinical_text, facecolor="#7ED957", fontsize=13)
add_box(ax, 0.92, 0.28, 0.07, 0.44, trust_text,   facecolor="#FF5C5C", fontsize=11)

# ---------- arrows ----------
add_arrow(ax, 0.24, 0.50, 0.27, 0.50)               # data -> core
add_arrow(ax, 0.45, 0.50, 0.49, 0.50)               # core -> models
add_arrow(ax, 0.69, 0.50, 0.72, 0.50)               # models -> clinical
add_arrow(ax, 0.90, 0.50, 0.92, 0.50)               # clinical -> trust

# continuous improvement feedback (bottom long arrow)
add_arrow(ax, 0.92, 0.18, 0.27, 0.18, text="Continuous Improvement", fontsize=13)

# ---------- export (VERY important) ----------
# Save high-quality PNG + PDF (PDF stays sharp when zooming)
plt.savefig("endocrine_big_picture.png", dpi=300, bbox_inches="tight")
plt.savefig("endocrine_big_picture.pdf", bbox_inches="tight")

plt.show()

print("Saved: endocrine_big_picture.png and endocrine_big_picture.pdf")
```
